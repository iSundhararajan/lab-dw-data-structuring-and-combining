```python
# Your code goes here
import numpy as np
import pandas as pd 
import statistics
url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_1 = pd.read_csv(url_1).squeeze("columns") 
```

```python
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
insurance_2 = pd.read_csv(url_2).squeeze("columns") 
```

```python
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
insurance_3 = pd.read_csv(url_3).squeeze("columns")
```

```python
display(insurance_1)
display(insurance_2)
display(insurance_3)
```

```python
insurance_1.columns = [col.lower() for col in insurance_1.columns]
insurance_2.columns = [col.lower() for col in insurance_2.columns]
insurance_3.columns = [col.lower() for col in insurance_3.columns]
```

```python
insurance_1.columns = insurance_1.columns.str.replace(' ', '_')
insurance_2.columns = insurance_2.columns.str.replace(' ', '_')
insurance_3.columns = insurance_3.columns.str.replace(' ', '_')
```

```python
insurance_1.rename(columns={'st': 'state'}, inplace=True)
insurance_2.rename(columns={'st': 'state'}, inplace=True)
insurance_3.rename(columns={'st': 'state'}, inplace=True)
```

```python
insurance_1['customer_lifetime_value'] = insurance_1['customer_lifetime_value'].str.replace('%', '')
insurance_1["customer_lifetime_value"]=insurance_1["customer_lifetime_value"].astype(float)/100
insurance_2['customer_lifetime_value'] = insurance_2['customer_lifetime_value'].str.replace('%', '')
insurance_2["customer_lifetime_value"]=insurance_2["customer_lifetime_value"].astype(float)/100
```

```python
complaints_1_mapping = {'1/0/00': 0, '1/2/00': 2,'1/1/00': 1,'1/3/00':3,'1/5/00':5,'1/4/00':4}
insurance_1['number_of_open_complaints'] = insurance_1['number_of_open_complaints'].map(complaints_1_mapping)
```

```python
insurance_1=insurance_1.dropna(how='all')
insurance_2=insurance_2.dropna(how='all')
insurance_3=insurance_3.dropna(how='all')
```

```python
vehicle_class_2_mapping = {"Sports Car": "Luxury", "Luxury SUV": "Luxury","Luxury Car": "Luxury",'Four-Door Car':'Four-Door Car','Two-Door Car':'Two-Door Car','SUV':'SUV'}
insurance_2['vehicle_class'] = insurance_2['vehicle_class'].map(vehicle_class_2_mapping)
```

```python
state_1_mapping = {'Arizona': 'Arizona', 'Washington': 'Washington','Nevada': 'Nevada','California':'California','Oregon':'Oregon','Cali':'California','AZ':'Arizona','WA':'Washington'}
insurance_1['state'] = insurance_1['state'].map(state_1_mapping)
```

```python
state_2_mapping = {'Arizona': 'Arizona', 'Washington': 'Washington','Nevada': 'Nevada','California':'California','Oregon':'Oregon','Cali':'California','AZ':'Arizona','WA':'Washington'}
insurance_2['state'] = insurance_2['state'].map(state_2_mapping)
```

```python
gender_1_mapping = {"F": "F", 'Femal': "F",'female': "F",'Male':'M','M':'M'}
insurance_1['gender'] = insurance_1['gender'].map(gender_1_mapping)
```
